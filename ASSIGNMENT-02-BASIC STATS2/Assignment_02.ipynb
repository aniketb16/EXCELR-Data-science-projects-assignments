```python
import pandas as pd
import numpy
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# @title Confidence interval using sample standard deviation(t-distribution)
```

```python
df=pd.DataFrame([1.13,1.55,1.43,0.92,1.25,1.36,1.32,0.85,1.07,1.48,1.20,1.33,1.18,1.22,1.29])
```

```python
df
```

```python
df.columns=['durability']
```

```python
df
```

```python
df.std()
```

```python
df.mean()
```

```python
df.describe()
```

```python
df.hist()
```

```python
df.boxplot()
```

```python
df_confidence=stats.t.interval(0.99,len(df)-1,loc=df.mean(),scale=df.std())
```

```python
df_confidence
```

```python
#so our confidence interval for 99%  will be from 0.6636 to 1.8136
```

```python
# @title Confidence interval using known population standard deviation
```

```python
df_population=stats.norm.interval(0.99,loc=df.mean(),scale=0.2)
```

```python
df_population
```

```python
#confidence interval if we know population standard deviation will be 0.7235 to 1.7538
```
