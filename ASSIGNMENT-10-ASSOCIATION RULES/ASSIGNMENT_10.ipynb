```python
def txt_split(txt):      #split text data in individual text by using txtsplt function
  return txt.split(',')
```

```python
data=df['Products'].apply(txt_split)
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
df_encoded=te.fit_transform(data)
```

```python
df_encoded
```

```python
te.columns_
```

```python
data=pd.DataFrame(df_encoded,columns=te.columns_)  #creating data frame
```

```python
data
```

```python
data.shape
```

```python
data.replace([True,False],[1,0],inplace=True)   #converting true false value to 1 and 0
```

```python
data.head()
```

```python
pd.set_option('display.max_columns',None)  #this shows all the columns
data[data.duplicated()]   #will only show duplicated rows
```

```python
data.describe()
```

```python
#lets make data more useful by removing very frequent items and very less frequent items as it will create baises and will dominate the analysis

series=data.sum()
```

```python
series.sort_values()
```

```python
series[(series >= 7500*0.02) & (series <= 7500*0.80)] #removing items with less than 2% and more than 80%
```

```python
new_data = data[series[(series >= 7500*0.02) & (series <= 7500*0.80)].keys().tolist()]
```

```python
new_data
```

```python
# @title APPLYING ASSOCIATION RULE

```

```python
frequent_itemsets=apriori(new_data,min_support=0.001,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets)
rules
```

```python
rules.sort_values('lift',ascending=False)
```
